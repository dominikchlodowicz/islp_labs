```python
import numpy as np

# np.array is a multidimensional array
x = np.array([[1, 2], [3, 4]])

# this shows number of dimensions of the array
x.ndim

# this show type of ints
x.dtype

# this shows docs for the chosen method
#np.array?

#here is how we can cast the data type in np array
np.array ([[1, 2], [3, 4]], float).dtype

#number of rows and columns 
x.shape

x.sum()

np.sum(x)

# we can reshape the array to shape of our choosing
x = np.array ([1, 2, 3, 4, 5, 6])
print('beginning x:\n', x)
x_reshape = x.reshape ((2, 3))
print('reshaped x:\n', x_reshape)


'''Operations in numpy are row-major ordered, and everything is order from 0 like whole python in general'''

# square root of np array
np.sqrt(x)

#generating random numbers
x = np.random.normal(size=50)
x

# loc - mean, scale - standard deviation
y = x + np.random.normal(loc=50, scale=1, size=50)
y

# calculating  correlation between matrixes
np.corrcoef(x, y)

# setting the random generation seed
rng = np.random.default_rng(1303)

# mean, variance, standard distribution
np.mean(x)
np.var(x)
np.std(x) 
```

```python
# Sequences

# seq of numbers
seq1 = np.linspace(0, 10, 11)
seq1

# seq of numbers spaced out by step
seq2 = np.arange(0, 10)
seq2
```

```python
'''In matplotlib, a plot consists of a figure and one or more axes. You can think of the figure as the
blank canvas upon which one or more plots will be displayed: 
it is the entire plotting window. The axes contain important
information about each plot, such as its x- and y-axis labels, title, and
more.'''

# subplots - the function returnsa tuple of length two: a figure object as well as the relevant axes object.
from matplotlib.pyplot import subplots
# this figsize argument is a size in inches
fig , ax = subplots(figsize =(8, 8))
x = rng.standard_normal (100)
y = rng.standard_normal (100)
ax.plot(x, y)
```

```python
fig , ax = subplots(figsize =(8, 8))
x = rng.standard_normal (100)
y = rng.standard_normal (100)
# creation of scatterplot, semicolon there makes matplot lib not returning text output
ax.scatter(x, y, marker='o');
```

```python
ig , ax = subplots(figsize =(8, 8))
ax.scatter(x, y, marker='o')
ax.set_xlabel("this is the x-axis")
ax.set_ylabel("this is the y-axis")
ax.set_title("Plot of X vs Y");
```

```python
# creation of multiple subplots, sharex - to share the same x axis etc.
fig , axes = subplots(nrows=2,
ncols=3,
figsize =(15, 5))
axes [0 ,1]. plot(x, y, 'o')
axes [1 ,2]. scatter(x, y, marker='+');
```
